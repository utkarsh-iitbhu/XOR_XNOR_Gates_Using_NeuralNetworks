```python
# Import Basic Libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')

# Import Libraries to construct Neural Network 
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
import numpy as np 
```

```python
# Initialising Data for our Model
X = np.array([[0,0],[0,1],[1,0],[1,1]])
yXOR = np.array([[0],[1],[1],[0]])
yXNOR = np.array([[1],[0],[0],[1]])
```

```python
# Building a Neural Network 
model = Sequential()

# Adding Layers to our Network 
model.add(Dense(8, input_dim=2))
model.add(Activation('sigmoid'))

# Output Layer which classifies 0 or 1 as Binary Output
model.add(Dense(1))
model.add(Activation('sigmoid'))
# Using Sigmoid Function i.e sigmoid = 1/(1+exp(-x))

# Choosing Optimiser as SGD and Learning Rate as 0.1
sgd = SGD(lr=0.1)

# Compiling our model and using binary_crossentropy as loss function for binary classification
model.compile(loss='binary_crossentropy', optimizer=sgd)
```

```python
# Input number 0 or 1 for different Gate
n = input("Enter a number 0 for XOR and 1 for XNOR : ")
```

```python
# Selecting the choosen path 
if n == '0':
    model.fit(X, yXOR, batch_size=1, epochs=1000)
    print(np.round(model.predict_proba(X)))      

if n=='1':
    model.fit(X, yXNOR, batch_size=1, epochs=1000)
    print(np.round(model.predict_proba(X)))   
```
